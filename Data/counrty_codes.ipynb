```python
import pandas as pd
```

Country codes web scraping

```python
url = "https://countrycode.org/"
```

```python
tables = pd.read_html(url)
tables
```

```python
counrty_df = tables[0]
counrty_df.columns = ['COUNTRY', 'COUNTRY CODE', 'ISO CODES', 'POPULATION', 'AREA KM2', 'GDP $USD']
counrty_df.head()
```

```python
counrty_df = counrty_df.rename(columns = {"COUNTRY":"Country"})
counrty_df = counrty_df.drop(columns=["COUNTRY CODE"])
counrty_df.head()
```

```python
counrty_df[["country_code", "Country_Code"]] = counrty_df["ISO CODES"].str.split('/', expand=True)
counrty_df.head()
```

```python
counrty_df = counrty_df.drop(columns=["ISO CODES"])
counrty_df.head()
```

```python
counrty_df.to_csv("Output_data/country_codes.csv")
counrty_df.head()
```
