```python
# Dependencie
import pandas as pd
```

```python
# Load in file
corruption_file1 = "corruption_final_version.csv"
# Read and display the CSV with Pandas
corruption_df1 = pd.read_csv(corruption_file1)
del corruption_df1 ["Unnamed: 0"]
corruption_df1.head()
```

```python
corruption_file2 = "corruption2016_2018.csv"
corruption_df2 = pd.read_csv(corruption_file2)
del corruption_df2 ["Country"]
corruption_df2.head()
```

```python
corruption_df2 = corruption_df2.rename(columns = {"ISO3":"Country_Code"})
corruption_df2 = corruption_df2[["Country_Code", "2016", "2017", "2018"]]
corruption_df2.head()
```

```python
# final_corruption_df = corruption_df1.merge(corruption_df2, how="left", on ="Country_Code")
# final_corruption_df
```

```python
final_corruption_df = pd.merge(corruption_df1, corruption_df2, how="outer", on="Country_Code")
final_corruption_df.head()
```

```python
corruption_df2.Country_Code = corruption_df2.Country_Code.str.strip()
corruption_df2.head()
```
