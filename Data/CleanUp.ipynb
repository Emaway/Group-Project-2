```python
import pandas as pd
import numpy as np
```

# Life expectancy at birth, male (years)

```python
# The path to our CSV file
file = "Life expectancy at birth, male (years).csv"

# Read our data into pandas
expectancyM_df = pd.read_csv(file, encoding = 'unicode_escape')
expectancyM_df.head()
```

```python
# Identify incomplete rows
expectancyM_df.count()
```

```python
# Replace NaN with '0'
expectancyM_df.replace(np.nan, 0, inplace=True)
expectancyM_df.head()
```

```python
# Import to csv
expectancyM_df.to_csv(r'LifeExpectancyMale.csv')
```
