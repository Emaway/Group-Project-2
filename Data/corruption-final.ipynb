```python
# Dependencie
import pandas as pd
```

```python
# Load in file
corruption_file = "corruption_Version_two.csv"
```

```python
# Read and display the CSV with Pandas
corruption_df = pd.read_csv(corruption_file)
del corruption_df["Unnamed: 0"]
corruption_df.head()
```

```python
country_file = "country_codes.csv"
country_codes_df = pd.read_csv(country_file)
del country_codes_df["Unnamed: 0"]
del country_codes_df["POPULATION"]
del country_codes_df["AREA KM2"]
del country_codes_df["GDP $USD"]
del country_codes_df["country_code"]
country_codes_df.head()
```

```python
corruption_df = corruption_df.merge(country_codes_df, how="inner", on =["Country", "Country"])
corruption_df.head()
```

```python
corruption_df = corruption_df[["Country", "Country_Code", "1998", "1999", "2000", "2001", "2002", "2003", "2004",
                               "2005", "2006", "2007", "2008", "2009", "2010", "2011", "2012", "2013", "2014", "2015"]]
corruption_df.head()
```

```python
corruption_df.to_csv("Output_data/corruption_final_version.csv")
corruption_df.head()
```
