```python
# Dependencie
import pandas as pd
```

```python
# Load in file
md_file = "Medical_doctors_per_10000.csv"
# Read and display the CSV with Pandas
medical_doctors_df = pd.read_csv(md_file)
medical_doctors_df.head()
```

```python
country_file = "country_codes.csv"
country_codes_df = pd.read_csv(country_file)
country_codes_df.head()
```

```python
doctors_df = medical_doctors_df.merge(country_codes_df, how="inner", on =["Country", "Country"])
del doctors_df["Unnamed: 0"]
del doctors_df["POPULATION"]
del doctors_df["AREA KM2"]
del doctors_df["GDP $USD"]
del doctors_df["country_code"]
doctors_df =  doctors_df[["Country", "Country_Code", "Year", "Medical doctors (number)", "Medical doctors (per 10 000 population)"]]
doctors_df.head()
```

```python
doctors_df.to_csv("Output_data/md_per_10000.csv")
doctors_df.head()
```
